Q3. You are given a list that contains some duplicate items. Remove the duplicates by converting the list to a set. Final output should be in list

shopping_list = ["apple", "banana", "apple", "orange", "banana", "grape"]

Expected Output: ['apple', 'grape', 'banana', 'orange']

```python
# Initial shopping list with duplicate items
shopping_list = ["apple", "banana", "apple", "orange", "banana", "grape"]

# 1. Convert the list to a set to remove duplicates
unique_items_set = set(shopping_list)

# 2. Convert the set back into a list
unique_shopping_list = list(unique_items_set)

# Display the final output
print(unique_shopping_list)
```

Q4. You are managing a list of students enrolled in two different courses. You need to perform various set operations to understand the student enrollment. Task: Create two sets: one for students in "Course A" and one for students in "Course B". Find the students who are in "Course A" but not in "Course B" (difference). Find students who are only in one of the two courses (symmetric difference). course_a = {"John", "Alice", "Bob", "David"} course_b = {"Alice", "Eve", "Charlie", "David"}

Using set methods

• Find students who are in Course A but not in Course B (difference)

• Find students who are only in one of the two courses (symmetric difference)

Expected Output:

Students in Course A but not in Course B: {'Bob', 'John'}

Students only in one of the two courses: {'Eve', 'John', 'Bob', 'Charlie'}

```python
# Initial student sets for both courses
course_a = {"John", "Alice", "Bob", "David"}
course_b = {"Alice", "Eve", "Charlie", "David"}

# 1. Find students in Course A but not in Course B (Difference)
only_course_a = course_a.difference(course_b)

# 2. Find students who are only in one of the two courses (Symmetric Difference)
unique_to_one_course = course_a.symmetric_difference(course_b)

# Display the results
print(f"Students in Course A but not in Course B: {only_course_a}")
print(f"Students only in one of the two courses: {unique_to_one_course}")
```

Q7. inventory = { &quot;The Great Gatsby&quot;: {&quot;price&quot;: 10.99, &quot;quantity&quot;: 5}, &quot;Rich dad poor dad&quot;: {&quot;price&quot;:
8.99, &quot;quantity&quot;: 8}, &quot;Harry Potter&quot;: {&quot;price&quot;: 12.50, &quot;quantity&quot;: 2}, }

• Add a new book in inventory using method and get all book titles using method

• Remove a book if it&#39;s out of stock

```python
# Initial bookstore inventory
inventory = { 
    "The Great Gatsby": {"price": 10.99, "quantity": 5}, 
    "Rich dad poor dad": {"price": 8.99, "quantity": 8}, 
    "Harry Potter": {"price": 12.50, "quantity": 2} 
}
```

```python
# 1. Add a new book using the .update() method
inventory.update({"To Kill a Mockingbird": {"price": 7.99, "quantity": 0}})
```

```python
# 2. Get all book titles using the .keys() method
book_titles = list(inventory.keys())
print("All Book Titles:", book_titles)
```

```python
# 3. Remove a book if it's out of stock (quantity == 0)
# We use list(inventory.items()) to safely modify the dictionary during the loop
for book, details in list(inventory.items()):
    if details["quantity"] == 0:
        inventory.pop(book)
```

```python
print(f"\nRemoved '{book}' because it was out of stock.")
```

```python
# Display final updated inventory
print("\nUpdated Inventory:",inventory)
```

Q10. Set Comparison
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

Find common elements.
Find elements only in Set A.
Find elements only in Set B.

```python
# Initial sets
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}
```

```python
# 1. Find common elements (Intersection)
common = A.intersection(B)
```

```python
# 2. Find elements only in Set A (Difference: A - B)
only_A = A.difference(B)
```

```python
# 3. Find elements only in Set B (Difference: B - A)
only_B = B.difference(A)
```

```python
# Display results
print(f"Common elements: {common}")
print(f"Elements only in Set A: {only_A}")
print(f"Elements only in Set B: {only_B}")
```
