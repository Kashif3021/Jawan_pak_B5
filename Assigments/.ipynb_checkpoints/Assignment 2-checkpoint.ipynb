#### Question 5
##### Create the following nested list and display:
##### students = [["Ali", 20], ["Ayesha", 22], ["Ahmed", 21]]

##### Ayesha


```python
students = [["Ali", 20], ["Ayesha", 22], ["Ahmed", 21]]
##For Ali
print(students[0][0])
##For Ayesha
print(students[1][0])
```

#### Question 6
##### Given list:
##### numbers = [10, 20, 30, 40, 50, 60, 70]
##### Display:
##### First 3 elements
##### Last 3 elements

```python
numbers = [10, 20, 30, 40, 50, 60, 70]
##First 3 elements
print(numbers[0:3])
##Last 3 elements
print(numbers[4:])
```
